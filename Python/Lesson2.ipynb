Взаимодействие с коллегами происходит с пулреквест и обучение пайтану тоже через пулреквест. 
Хэлперпул- это форк в котором мы вносим комментарии, где непонятно и отправляем пулреквест.
Для того, чтобы сделать хэлперпул: 1) сделать форк 2) клонирование в кошке 3) внести комментарии, где непонятно 4) отдать коммит и пулреквест 5) написать в чатик в гидхаб: "дай ревью на хэлперпул" + ссылка на пулреквест +@отметить лицо
